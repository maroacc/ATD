```python
# Imports
import csv
```

```python
def load_csv(file_path):
    with open(file_path, "r") as input_file:
        reader = csv.reader(input_file, delimiter = ";")
        matrix = [row for row in reader]
    return matrix
```

```python
matrix = load_csv("tarifas_electricas.csv")
```
