```python
 # Inventario de estaciones

def inventario_estaciones():
    """Realiza una petición para obtener el inventario de todas las estaciones de la AMET

    Returns:
        json_dict: Diccionario json con el inventario de todas las estaciones de la AMET (e.g., 'dataset/training').
    """
    # Petición AMET para obtener la url al inventario estaciones
    url = "https://opendata.aemet.es/opendata/api/valores/climatologicos/inventarioestaciones/todasestaciones/"
    querystring = {"api_key": api_key}
    headers = {
    'cache-control': "no-cache"
    }
    response = requests.request("GET", url, headers=headers, params=querystring)
    url = response.json()["datos"]
    headers = {
    'cache-control': "no-cache"
    }
    # Petición para obtener el inventario de estaciones
    response = requests.request("GET", url, headers=headers)
    return response.json()
```

```python
def get_climatologia(idema, fecha_ini, fecha_fin):
    """Realiza una petición para obtener la climatología diaria de una estación en un periodo.
    
    Args:
        idema: idema de la estación (e.g., '3194U').
        fecha_ini: fecha de inicio del periodo (e.g., '2019-10-01T00:00:00UTC').
        fecha_fin: fecha de fin del periodo (e.g., '2019-10-30T00:00:00UTC')

    Returns:
        json_dict: Diccionario json con la climatología diaria de la estación en el periodo especificado.
    """
    url = f'https://opendata.aemet.es/opendata/api/valores/climatologicos/diarios/datos/fechaini/{fecha_ini}/fechafin/{fecha_fin}/estacion/{idema}'
    querystring = {"api_key": api_key}
    headers = {
    'cache-control': "no-cache"
    }
    response = requests.request("GET", url, headers=headers, params=querystring)
    json_dict = response.json()
    url = json_dict["datos"]
    headers = {
    'cache-control': "no-cache"
    }
    response = requests.request("GET", url, headers=headers)
    json_dict = response.json()
    return json_dict
```

```python
# Params
year_ini = 2011
year_fin = 2020
```

```python
def get_climatologia_todas_estaciones(fecha_ini, fecha_fin):
    """Realiza una petición para obtener la climatología diaria de todas las estaciones en un periodo.
    
    Args:
        fecha_ini: fecha de inicio del periodo (e.g., '2019-10-01T00:00:00UTC').
        fecha_fin: fecha de fin del periodo (e.g., '2019-10-30T00:00:00UTC')

    Returns:
        json_dict: Diccionario json con la climatología diaria de todas las estaciones en el periodo especificado.
    """
    url = f"https://opendata.aemet.es/opendata/api/valores/climatologicos/diarios/datos/fechaini/{fecha_ini}-08-01T00%3A00%3A00UTC/fechafin/{fecha_fin}-08-30T23%3A59%3A59UTC/todasestaciones"
    querystring = {"api_key": api_key}
    headers = {
        'cache-control': "no-cache"
        }
    response = requests.request("GET", url, headers=headers, params=querystring)
    json_dict = response.json()
    url = json_dict["datos"]
    headers = {
        'cache-control': "no-cache"
        }
    response = requests.request("GET", url, headers=headers)

    return response.json()
```

```python
def calcular_media(dict_json):
    """Calcula la temperatura media de las estaciones dadas.
    
    Args:
        dict_json: json con climatologias diarias de estaciones.

    Returns:
        t_med: temperatura media de las estaciones.
    """
    acc = 0
    count = 0
    for keyval in dict_json:
        if keyval.get("tmed")!= None:
            acc = acc + float(keyval.get("tmed").replace(",", "."))
            count+=1
    t_med = acc/count
    return t_med
```

```python
# Itermamos sobre get_climatologia_todas_estaciones() 
# para obtener la climatología de cada año de la década
n_years = year_fin - year_ini + 1
temp_med_nac = np.zeros(n_years)

for year in range(n_years):
    fecha_ini = f"{year_ini + year}-08-01T00%3A00%3A00UTC"
    fecha_fin = f"{year_ini + year}-08-30T00%3A00%3A00UTC"
    climat = get_climatologia_todas_estaciones(fecha_ini, fecha_fin)
    temp_med_nac[year] = calcular_media(climat)

print(temp_med_nac)
```
